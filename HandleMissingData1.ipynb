#### Make the day column a date

```python
dateparse = lambda x: pd.datetime.strptime(x, '%d/%m/%Y') #01/01/2018
df = pd.read_csv('resources/WeatherData.csv', parse_dates=['day'], date_parser=dateparse)
type(df.day[0])
```

#### Set the day column as the dataframe index

```python
df = df.set_index('day')
df
```

### Handling Missing data 

### fillna

```python
new_df = df.fillna(0)
new_df
```

#### fillna sparingly

```python
new_df = df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'No Event'
})
new_df
```

#### forward fillna

```python
new_df = df.fillna(method='ffill')
new_df
```

#### backward fillna

```python
new_df =df.fillna(method='bfill')
new_df
```

#### ffill with limit

```python
new_df = df.fillna(method='ffill', limit=1)
new_df
```

### Interpolate 

```python
new_df = df.interpolate()
new_df
```

#### Interpolate with time

```python
new_df = df.interpolate(method='time')
new_df
```
